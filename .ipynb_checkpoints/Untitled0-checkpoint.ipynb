```python
import seaborn as sns
import scipy 
#import numpy as np
import pandas as pd
%matplotlib inline
import psycopg2
import matplotlib.pyplot as plt
from ggplot import *
```

```python
data = pd.read_csv(r'H:\notebooks\testdata.csv')
data['Group'] = data["Crop"].map(str) + " " + data["Irrifation"].map(str) 
grouped_data = data.groupby(['Group', 'Time', 'Plot', 'Depth'])['Value'].mean()
new_group=pd.DataFrame(grouped_data)
new_data=new_group.reset_index()
new_data
```

```python
sns.set(font_scale=0.5)
#pal = sns.hls_palette(8, l=.3, s=.8)
trial_plot= sns.factorplot('Time', 'Value', 'Group', new_data, row='Depth', kind='point', size=15, aspect=20, sharey=False, sharex=False)
```

```python
grouped_data = new_data.groupby(['Group', 'Time', 'Plot'])['Value'].mean()
new_group=pd.DataFrame(grouped_data)
new_group.reset_index(inplace=True)
pal = sns.hls_palette(8, l=.3, s=.8)
rainplot = sns.tsplot(new_group, time="Time", unit="Plot", condition="Group", value="Value", linewidth=2.0, color=pal, ci=80)
```
