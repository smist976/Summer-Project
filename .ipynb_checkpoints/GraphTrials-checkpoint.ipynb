```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import psycopg2
from ggplot import * 
import seaborn as sns
```

```python
start_date = "2014-11-25"
end_date= "2014-12-01"
sql_select = "SELECT b.measuretype, b.cultivar , b.irrigation, b.depth, a.time, AVG (a.measurement) AS average from alphasensordata a  INNER JOIN alphasensorinfo b  ON a.sensorid=b.sensorid WHERE a.sensorid=b.sensorid AND a.time BETWEEN '" + start_date +"' and '"+ end_date + "'GROUP BY b.measuretype, b.cultivar , b.irrigation, b.depth, a.time"  
dbconn=psycopg2.connect(host="aklpdb22.pfr.co.nz",user="rainshelterwrite", password="luverly Jubbly", database="rainshelter")
dbcursor = dbconn.cursor()
dbcursor.execute(sql_select)
data = dbcursor.fetchall()
names = dbcursor.description
dbconn.commit()
dbconn.close()
```

```python
from ggplot import meat
meat_lng = pd.melt(meat, id_vars=['date'])
ggplot(aes(x='date', y='value', colour='variable'), data=meat_lng) + geom_line()
```

```python
columnnames = [column[0] for column in names]
df = pd.DataFrame(data, columns=columnnames)
#df2=df.set_index(['measuretype'])
df2= df.query('measuretype=="ElectricalConductivity"')
df2["Group"] = df2["cultivar"].map(str) + df2["irrigation"].map(str) + " " + df2['depth'] 
#df2["Group2"] = df2["Group"] = df2["cultivar"].map(str) + df2["irrigation"].map(str)
df2
```

```python
ggplot(aes(x='time', y='average', color='Group'), df2) +\
           geom_line() 
```

```python
grouped_data = df2.groupby('depth')
#grouped_data = df2.groupby(level=['depth'],axis=1).mean()
grouped_data
for name, group in grouped_data:
        print (name)
```

```python
ggplot(aes(x='time', y='average', colour='Group'), df2) + geom_line() + facet_wrap("depth")+ \
scale_x_date(breaks = date_breaks('1 day'), labels= date_format("%b %d %Y"))
```
