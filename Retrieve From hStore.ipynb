```python
import psycopg2
import pandas as pd
```

```python
#Define sensor data required
measure_type="VolumetricWaterContent"

#retrieved sensor info from database 
sql_statement="Select sensorid, cultivar, irrigation, depth, plot from alphasensorinfo where measuretype='" + measure_type + "'"
dbconn=psycopg2.connect(host="aklpdb22.pfr.co.nz",user="rainshelterwrite", password="luverly Jubbly", database="rainshelter")
dbcursor = dbconn.cursor()
dbcursor.execute(sql_statement)
sensor_info=dbcursor.fetchall()
column_names = dbcursor.description
dbconn.commit()
dbconn.close()
```

```python
# create data frame with sensor information
columnnames = [column[0] for column in column_names]
sensorinfo = pd.DataFrame(sensor_info, columns=columnnames)
```

```python
#retrieve sensor id from database 
sql_statement="Select sensorid from alphasensorinfo where measuretype='" + measure_type + "'"
dbconn=psycopg2.connect(host="aklpdb22.pfr.co.nz",user="rainshelterwrite", password="luverly Jubbly", database="rainshelter")
dbcursor = dbconn.cursor()
dbcursor.execute(sql_statement)
sensor_id=dbcursor.fetchall()
dbconn.commit()
dbconn.close()
```

```python
# retrieve values relating to sensor ids list
all_items_sql =""
start_date = "2014-11-05"
end_date = "2014-11-30"


for item in sensor_id:
    item = str(item)[1:-2]
    text_item = item.replace ('\'', '"')
    item_sql = ", data -> " + item + " as "+ text_item  
    all_items_sql = all_items_sql + item_sql

retrieve_data_sql = "Select time as time " + all_items_sql + " from hstorealpha where time between '" + start_date + "' and '" + end_date + "'"
```

```python
# retrieve measurement values
dbconn=psycopg2.connect(host="aklpdb22.pfr.co.nz",user="rainshelterwrite", password="luverly Jubbly", database="rainshelter")
dbcursor = dbconn.cursor()
dbcursor.execute(retrieve_data_sql )
all_data=dbcursor.fetchall()
data_columns = dbcursor.description
dbconn.commit()
dbconn.close()
```

```python
# create data frame of measurement values
data_column_names = [column[0] for column in data_columns]
sensor_data = pd.DataFrame(all_data, columns=data_column_names)
```

```python
sensor_data.set_index('time', inplace=True)
```

```python
stacked_data = sensor_data.stack()
long_sensor_data = pd.DataFrame(stacked_data)
long_sensor_data.reset_index(inplace=True)
long_sensor_data.columns=['time', 'sensorid', 'value']
```

```python
complete_data = pd.merge(long_sensor_data, sensorinfo, on ='sensorid', how='inner')
```
