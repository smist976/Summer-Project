```python
import seaborn as sns
import scipy 
import numpy as np
import pandas as pd
%matplotlib inline
import psycopg2
import matplotlib.pyplot as plt
from ggplot import *
```

### Example of time series from Seaborn website http://stanford.edu/~mwaskom/software/seaborn/examples/timeseries_from_dataframe.html

```python
#create data set for plot
def gamma_pdf(x, shape, coef, obs_err_sd=.1, tp_err_sd=.001):
    y = scipy.stats.gamma(shape).pdf(x) * coef
    y += np.random.normal(0, obs_err_sd, 1)
    y += np.random.normal(0, tp_err_sd, len(x))
    return y

gammas = []
n_units = 20
params = [(5, 1), (8, -.5)]
x = np.linspace(0, 15, 31)
for s in range(n_units):
    for p, (shape, coef) in enumerate(params):
        y = gamma_pdf(x, shape, coef)
        gammas.append(pd.DataFrame(dict(condition=[["pos", "neg"][p]] * len(x),
                                        subj=["subj%d" % s] * len(x),
                                        time=x * 2,
                                        BOLD=y), dtype=np.float))
gammas = pd.concat(gammas)
```

```python
#data set 
gammas
```

```python
# create polt with one line of code
plot = sns.tsplot(gammas, time="time", unit="subj", condition="condition", value="BOLD")
```

### Example Using Volumetric Water Content Data from RainShelter Trial

```python
#Retrieve data from the database between two dates. 
start_date = "2014-11-05"
end_date= "2014-11-30"
sql_select = "SELECT a.measurement as measuretype, a.cultivar, a.irrigation, a.plot, a.depth, b.time, b.measurement FROM alphasensorinfo a INNER JOIN alphasensordata b ON a.sensorid=b.sensorid WHERE b.time BETWEEN '" + start_date +"' and '"+ end_date + "'"  
dbconn=psycopg2.connect(host="aklpdb22.pfr.co.nz",user="rainshelterwrite", password="luverly Jubbly", database="rainshelter")
dbcursor = dbconn.cursor()
dbcursor.execute(sql_select)
data = dbcursor.fetchall()
names = dbcursor.description
dbconn.commit()
dbconn.close()
```

The following cells calculate soil water deficit value from the volumetric water content sensor and convert data into format for plotting

```python
#put data into Dataframe 
columnnames = [column[0] for column in names]
AllData = pd.DataFrame(data, columns=columnnames)
```

```python
#select only Volumetric Water Content
WaterContentData= AllData.query('measuretype=="VolumetricWaterContent"')

#calculate layer water value for each depth_date
WaterContentData['LayerWater']=WaterContentData['measurement']*300
WaterContentData['LayerWater'][WaterContentData['depth']== 'D1I'] = WaterContentData['measurement'] * 150/2
WaterContentData['LayerWater'][WaterContentData['depth']== 'D1B'] = WaterContentData['measurement'] * 150/2
WaterContentData['LayerWater'][WaterContentData['depth']== 'D2'] = WaterContentData['measurement'] * 150

#create group  
WaterContentData['Group'] = WaterContentData["cultivar"].map(str) + " " + WaterContentData["irrigation"].map(str)

# group data frame
group_WCD=WaterContentData.groupby(['time', 'plot', 'Group'])['LayerWater'].sum()
ProfileWater=pd.DataFrame(group_WCD)
ProfileWater.rename(columns={'LayerWater':'ProfileWater'}, inplace=True)

ProfileWater.reset_index(inplace=True)
ProfileWater.set_index(pd.DatetimeIndex(ProfileWater['time']), inplace=True)
ProfileWater.index.name="index"

# select only data from 05/11/2014
x=pd.datetime(2014, 11, 5, 7, 45)#Time 7.45 had NAN values
y=pd.datetime(2014, 11, 5, 23, 45)
ProfileWater.sort(inplace=True)
fc_data = ProfileWater.ix[x:y]

# create another dataframe from 05/11/2014 data and group and select maxvalue and add 40 to it 
FieldCapacity = fc_data.groupby(['Group', 'plot'])['ProfileWater'].max() +40
FieldCapacityData=pd.DataFrame(FieldCapacity)
FieldCapacityData.rename(columns={'ProfileWater':'FieldCapacity'}, inplace=True)

FieldCapacityData.reset_index(inplace=True)
#change index on profile_water data frame
ProfileWater.reset_index(inplace=True)
```

```python
FieldCapacityData
```

```python
SoilWaterDeficit=pd.merge(ProfileWater,FieldCapacityData, on=['Group', 'plot'], how='inner', suffixes='fc')                                  
```

```python
SoilWaterDeficit['SoilWaterDeficit']= -(SoilWaterDeficit['FieldCapacity']-SoilWaterDeficit['ProfileWater'])
SoilWaterDeficit.drop('index', 1, inplace=True)
```

```python
#example of the data being plotted
SoilWaterDeficit
```

# Seaborn TS Plot

```python
sns.set_style("dark")
plt.rcParams['figure.figsize'] = 14, 12  # set size of figure
pal = sns.hls_palette(8, l=.3, s=.8)

rainplot = sns.tsplot(SoilWaterDeficit, time="time", unit="plot", condition="Group", value="SoilWaterDeficit", linewidth=2.0, color=pal, ci=80)

# to show traces of all remove ci=50 add err_style="unit_traces" - can also have both

rainplot.legend(loc=3,prop={'size':16})
plt.xlabel("Time", fontsize=18, fontweight="bold")
plt.ylabel("Soil Water Deficit (mm)", fontsize=18, fontweight="bold")
plt.xticks(fontsize= 16, fontweight="bold")
plt.yticks(fontsize= 16, fontweight="bold")
plt.title("Rain Shelter Barley 2014/15", fontsize=28, fontweight="roman")
```

Plot created with mat plot lib. uses seaborn graphics.

```python
SoilWaterDeficitData = SoilWaterDeficit.drop(['FieldCapacity', 'plot', 'ProfileWater'], axis=1)
```

```python
fig = plt.figure(figsize=(16,12))
ax = fig.add_subplot(111)


SoilWaterDeficitData.groupby(["time", "Group"]).mean().unstack("Group").plot(ax=ax, title="Rain Shelter Barley 2014/15")
#SoilWaterDeficitData.plot(ax=ax)

x=SoilWaterDeficitData.groupby(["Group", "time"]).mean().unstack("Group").index
palette=sns.color_palette()

for cond, cond_df in SoilWaterDeficitData.groupby("Group"):
    low = cond_df.groupby("time")["SoilWaterDeficit"].min() 
    high = cond_df.groupby("time")["SoilWaterDeficit"].max() 
    ax.fill_between(x, low, high, alpha=.2, color=palette.pop(0))

    
ax.grid(False)
ax.set_title(ax.get_title(), fontsize=28)
ax.set_ylabel("Soil Water Deficit (mm)", fontsize=14, fontweight="bold")
ax.set_xlabel("Date", fontsize=14, fontweight="bold")
ax.tick_params(labelsize=14, width=5)
ax.legend(loc=3, fontsize=16)

#mpl.to_bokeh(name="test")
```
