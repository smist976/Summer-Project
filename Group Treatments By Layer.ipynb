```python
start_date = "2014-12-01"
end_date= "2014-12-16"
sql_select = "SELECT a.measuretype, a.cultivar, a.irrigation, a.plot, a.depth, b.time, b.measurement FROM alphasensorinfo a INNER JOIN alphasensordata b ON a.sensorid=b.sensorid WHERE b.time BETWEEN '" + start_date +"' and '"+ end_date + "'"  
dbconn=psycopg2.connect(host="aklpdb22.pfr.co.nz",user="rainshelterwrite", password="luverly Jubbly", database="rainshelter")
dbcursor = dbconn.cursor()
dbcursor.execute(sql_select)
data = dbcursor.fetchall()
names = dbcursor.description
dbconn.commit()
dbconn.close()
```

```python
columnnames = [column[0] for column in names]
df = pd.DataFrame(data, columns=columnnames)
df2= df.query('measuretype=="VolumetricWaterContent"')
df2
```

```python
df2[Group'] = df2["cultivar"].map(str) + " " + df2["irrigation"].map(str)
df2['ProfileWater']= df['measurement'] * 300
df2['ProfileWater'][df['depth']== '0-15(InRow)'] = df['measurement'] * 150/2
df2['ProfileWater'][df['depth']== '0-15(btwRow)'] = df['measurement'] * 150/2
df2['ProfileWater'][df['depth']== '15-30'] = df['measurement'] * 150
df2['Group'] = df2["cultivar"].map(str) + " " + df2["irrigation"].map(str)
columns = ['measuretype', 'cultivar', 'irrigation', 'measurement']
new_data = df2.drop(columns, 1)
grouped_data = new_data.groupby(['Group', 'time', 'depth'])['ProfileWater'].mean()
new_group =pd.DataFrame(grouped_data)
new_data=new_group.reset_index()
new_data
```
